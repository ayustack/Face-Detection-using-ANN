```python
from matplotlib import pyplot as plt
import cv2
import sys
```

```python
cPath="D:\\Face Detection using ANN\\haarcascade_frontalface_alt.xml"
```

```python
pedsCascade = cv2.CascadeClassifier(cPath)
```

```python
image = cv2.imread("D:\\Face Detection using ANN\\peds.jpg")
RGB_im = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_im)
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
peds = pedsCascade.detectMultiScale(gray, scaleFactor = 1.2, minNeighbors = 1, minSize = (2,2))
```

```python
print("Found {} pedestrian!".format(len(peds)))
```

```python
for(x,y,w,h) in peds :
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),3)
```

```python
status = cv2.imwrite("pedsnew.jpg",image)
print("Image written to file system : ",status)
```
